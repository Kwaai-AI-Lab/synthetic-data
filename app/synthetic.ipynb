````python
import random
import datetime
import os

def generate_synthetic_emails(num_emails, example_email):
    emails = []
    for i in range(num_emails):
        sender_email = example_email['from']
        recipient_emails = ["Jeany@gmail.com"]
        subject = example_email['subject']
        date_sent = datetime.datetime.strptime(example_email['date'], "%a, %d %b %Y %H:%M:%S %z")
        body_text = example_email['body']
        attachments = ["attachment1.pdf", "attachment2.docx"]
        
        email = {
            'sender_email': sender_email,
            'recipient_emails': recipient_emails,
            'subject': subject,
            'date_sent': date_sent,
            'body_text': body_text,
            'attachments': attachments
        }
        emails.append(email)
    
    return emails

# Create folder if it doesn't exist
folder_name = "synthetic_emails"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Give examples of emails
example_email = {
    "subject": "OPay is not running any Christmas offer",
    "from": "OPay <no-reply@opay-us.com>",
    "date": "Wed, 20 Dec 2023 23:04:49 +0530",
    "body": "OPay is not giving out N15,000 or any other amount this Christmas. Beware of Fraudsters trying to defraud people of their hard earned money during this festive period! Signed: OPay MGT Click here to unsubscribe"
}

num_emails = 1000
synthetic_emails = generate_synthetic_emails(num_emails, example_email)

# Save each email to a separate file
for i, email in enumerate(synthetic_emails):
    filename = os.path.join(folder_name, f"email_{i}.txt")
    with open(filename, 'w') as file:
        file.write(f"Sender: {email['sender_email']}\n")
        file.write(f"Recipient(s): {', '.join(email['recipient_emails'])}\n")
        file.write(f"Subject: {email['subject']}\n")
        file.write(f"Date Sent: {email['date_sent'].strftime('%Y-%m-%d %H:%M:%S %Z')}\n")
        file.write("\n")
        file.write(email['body_text'])
    print(f"Email {i+1} saved to {filename}")
````

````python
# Use gpt-3.5 model to build the classifier model
import os
from openai import OpenAI

client = OpenAI(
    # This is the default and can be omitted
    api_key=os.environ.get("OPENAI_API_KEY"),
)

def chat_completion(prompt, model="gpt-3.5-turbo"):
    messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0,
)
    return response.choices[0].message.content
````

````python
prompt = f"""
You are a machine learning engineer with 10 years of experience, \
Your task is to classify the 'body_text' in the email text delimited by the three backticks with the following: \
Classification : \
   
Professional \
Personal \
Undefined \

Print the text in the body_text

email text: ```{email}```

"""

response = chat_completion(prompt)
print(response)
````
